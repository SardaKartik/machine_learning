```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df
```

```python
df.head()
```

```python
df.columns
```

```python
new=df.drop('Name',axis="columns")
```

```python
new.head()
```

```python
new=new.drop('PassengerId',axis='columns')
```

```python
new
```

```python
new=new.drop(['Ticket','Fare','Cabin','Embarked'],axis='columns')
```

```python
new.head()
```

```python
new
```

```python
inputs=new.drop('Survived',axis='columns')
```

```python
inputs.head()
```

```python
output=new['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex=LabelEncoder()
```

```python
inputs['sex_n']=le_sex.fit_transform(inputs['Sex'])
```

```python
inputs
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
```

```python
inputs
```

```python
inputs=inputs.fillna(int(inputs['Age'].mean()))
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,output,train_size=0.8)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_prediction=model.predict(x_test)
```

```python
y_prediction
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_prediction)
```

```python
cm
```

```python
import seaborn as sn
plt.figure(figsize=(5,4))
sn.heatmap(cm,annot=True)
plt.xlabel('prediction')
plt.ylabel('truth')
```
