```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),digits.data, digits.target, cv=5).mean()
```

```python
cross_val_score(RandomForestClassifier(n_estimators=10),digits.data,digits.target).mean()
```

```python
cross_val_score(LogisticRegression(),digits.data,digits.target).mean()
```

```python
cross_val_score(GaussianNB(),digits.data,digits.target).mean()
```

```python
cross_val_score(MultinomialNB(),digits.data,digits.target).mean()
```

```python
cross_val_score(DecisionTreeClassifier(),digits.data,digits.target).mean()
```
