```python
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
```

RANDOM FOREST
The basic idea behind this is to combine multiple decision trees in determining the final output rather than relying on individual decision trees. 

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.data[:5]
```

```python
df=pd.DataFrame(digits.data)
```

```python
df
```

```python
df['target']=digits.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(['target'],axis="columns"),df['target'],test_size=0.2,random_state=8)
```

```python
x_train
```

```python
len(x_train)
```

```python
y_train
```

```python
len(y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

ensemble term is used when we are using multiple algorithm 

```python
model=RandomForestClassifier()
```

here the n_estamator is wornking n=100 that why we are getting the accuracy - 98

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model1=RandomForestClassifier(n_estimators=1000)
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_train,y_train)
```

it is given 1.0 mean 100 percent because we are predicting the score of  the model on the same parameter or same input on which we trained the model 

```python
model.score(x_test,y_test)
```

```python
# cofusion matrix 
```

```python
y_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_predicted)
```

```python
cm
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(10,5))
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel("Truth")
```
