```python
import pandas as pd
```

```python
df=pd.read_csv("tested.csv")
```

```python
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df
```

```python
target=df.Survived
inputs=df.drop('Survived',axis="columns")
```

since machine learning algorithm works only on the number so we have convert males and females into number 

```python
dumies=pd.get_dummies(inputs.Sex)
```

```python
dumies.head(3)
```

```python
inputs=pd.concat([inputs,dumies],axis='columns')
```

```python
inputs
```

```python
inputs.drop(['Sex'],axis='columns',inplace=True)
```

```python
inputs
```

to find out whether there is any na value in the coloum 

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:15]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
# filling na value with mean
```

```python
inputs
```

```python
inputs.Fare=inputs.Fare.fillna(inputs.Fare.mean())
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
